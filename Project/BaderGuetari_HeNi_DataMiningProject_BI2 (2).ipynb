```python
#Although we already know that there are 3 types of labels, if we don't know the number of types of labels, we need to use some methods to determine the number of labels, i.e. the value of Kmean species K
```

```python
#We know that we should have 3 clusters (depend on the labels)
```

```python
#The elbow method
```

```python
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
distortions = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(DataPca)
    distortions.append(sum(np.min(cdist(DataPca, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

print("It looks like point 2 and point 3, both work, but it feels like point 2 has more curvature? If we go for point 2, then it doesn't fit.")
```

```python
#FMI score
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import  fowlkes_mallows_score
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=23).fit(DataPca)
    score=fowlkes_mallows_score(target,kmeans.labels_)
    print("The FMI score for the {}th clustering model is{}".format(i,score))

print("\nThe FMI score is also maximum for k=2, but it also does not match")
```

```python
#silhouette_score
```

```python
from sklearn.metrics import  silhouette_score
import matplotlib.pyplot as plt
silhouettteScore=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=23).fit(DataPca)
    score=silhouette_score(DataPca,kmeans.labels_)
    silhouettteScore.append(score)
plt.figure(figsize=(12,6))
plt.plot(range(2,11),silhouettteScore,linewidth=1.5,linestyle='-',color='red')
plt.show()

print("the silhouette_score from 3->4  is the most, so we choose k = 3")
```

```python
#calinski_harabasz_score
```

```python
from sklearn.metrics import calinski_harabasz_score
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,random_state=23).fit(DataPca)
    score=calinski_harabasz_score(DataPca,kmeans.labels_)
    print("The Calinski-Harabasz index score for the {}th clustering model is{}".format(i,score))

print("\nwhen k = 3 the Calinski-Harabasz is highest,and we have a reason to choose k = 3")
```

```python
plt.figure(figsize=(10, 8))
dendrogram(Z, truncate_mode='lastp', p=20, show_leaf_counts=False, leaf_rotation=90, leaf_font_size=15, show_contracted=True)
plt.title('Dendrogram for the Agglomerative Clustering')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.show()
```
